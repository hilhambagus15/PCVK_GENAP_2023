2. MERUBAH UKURAN

```python
img4 = cv.resize(cv.cvtColor(img,cv.COLOR_BGR2RGB),(512,1024))
plt.imshow(img4)
```

3. MEMBUAT GARIS VERTIKAL DAN HORIZONTAL

```python
img4 = cv.cvtColor(img, cv.COLOR_BGR2RGB)

tinggi = img.shape[0]
lebar = img.shape[1]

for x in range (tinggi):
  img4[x, int(lebar/2)] = [255, 255, 255]
for y in range (lebar):
  img4[int((tinggi)/2),y] = [255, 255, 255]

plt.imshow(img4)
```

4 FLIP VERTIKAL ATAU HORIZONTAL

```python
img = cv.imread('/content/drive/MyDrive/Pengolahan Citra/SMILEY.jpg')
img5 = cv.flip(img, 0)
img6 = cv.flip(img, 1)

fig = plt.figure(figsize=(20,10))
fig.add_subplot(1,2,1)
plt.imshow(img5)
plt.title("Vertical")
fig.add_subplot(1,2,2)
plt.imshow(img6)
plt.title("Horizontal")
```

5. MENAMBAH SHAPE

```python
img = cv.imread('/content/drive/MyDrive/Pengolahan Citra/SMILEY.jpg')
cv.rectangle(img,pt1=(25,70),pt2=(220,150),color=(255,0,0),thickness=8)
plt.imshow(img)
```
