```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img=cv2.imread('/content/drive/MyDrive/Pengolahan Citra/NOISE.jpg')
plt.figure(figsize=(5,5))
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
def conv(X, F):
    X_height, X_width = X.shape
    F_height, F_width = F.shape
    H = (F_height) // 2
    w = (F_width) // 2

    out = np.zeros((X_height, X_width))

    for i in np.arange(H+1, X_height - H):
        for j in np.arange(w+1, X_width - w):
            sum = 0
            for k in np.arange(-H, H):
                for l in np.arange(-w, w):
                    a = X[i+k, j+l]
                    filter_weight = F[H+k, w+l]
                    sum += (a * filter_weight)
            out[i, j] = sum

    return out
```

```python
img = cv2.imread('/content/drive/MyDrive/Pengolahan Citra/NOISE.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

mean = (1.0 / 9) * np.array(
    [[1, 1, 1],
     [1, 1, 1],
     [1, 1, 1]])

img_out = conv(img, mean)
plt.imshow(img_out, cmap='gray', interpolation='bicubic')
plt.axis('off')
plt.show()
```
